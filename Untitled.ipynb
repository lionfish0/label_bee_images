```python
import numpy as np
from numpy import unravel_index
from glob import glob
import pandas as pd
import png
from retrodetect import getblockmaxedimage, detect,ensemblegetshift,alignandsubtract
import matplotlib.pyplot as plt
%matplotlib inline

pathtoimgs = '/home/mike/Documents/Research/bee/photos_April5'
labels = pd.read_csv('labels.csv',header=None)
patchn = 0
savesize = 40
delsize=15
```

```python
def getimgfilename(number):
    fns = sorted(glob('%s/*.np'%(pathtoimgs)))
    return fns[number]
    
def addimg(imglist,n):
    _, img, data = np.load(getimgfilename(n),allow_pickle=True)
    if 'photoaverages' in data:
        if (data['photoaverages']['this']>data['photoaverages']['last']+1):
            if img is not None: imglist.append(img.astype(np.float))
        
patches = []

for l in labels.to_numpy()[12:101:4]:

    last_flash = []
    last_noflash = []
    this_flash = []
    this_noflash = []
    n =l[0]
    fn=l[1]
    truex =l[2]
    truey =l[3]
    print(n)
    assert getimgfilename(n)==fn
    
    _, img, data = np.load(getimgfilename(n-3),allow_pickle=True)
    #if (data['photoaverages']['this']>data['photoaverages']['last'])+3: lastn = n-3
    #else: lastn = n-6
    
    last_noflash = []
    last_flash = []
    for i,lastn in enumerate([n-3,n-6,n-9]):
        last_nof = []
        last_f = []
        addimg(last_nof,lastn-1)
        addimg(last_f,  lastn)
        addimg(last_nof,lastn+1)
        if (len(last_nof)>0) and (len(last_f)>0):
            last_noflash.append(last_nof)
            last_flash.append(last_f)
    addimg(this_noflash,n-1)
    addimg(this_flash,  n)
    addimg(this_noflash,n+1)

    
    if len(this_flash)==0: continue
    if len(this_noflash)==0: continue
        
    this_diff = None
    
    for nf in this_noflash: 
        diff = detect(this_flash[0],nf,blocksize=10)
        if this_diff is None:
            this_diff = diff
        else:
            this_diff = np.minimum(diff,this_diff)

    last_diff = None
    for i in range(len(last_flash)):
        if len(last_flash[i])==0: continue
        if len(last_noflash[i])==0: continue
        for nf in last_noflash[i]: 
            diff = detect(last_flash[i][0],nf,dilate=False)
            if last_diff is None:
                last_diff = diff
            else:
                last_diff = np.maximum(diff,last_diff)

    if last_diff is None: continue
    if this_diff is None: continue
    bg = this_noflash[0].copy()
    bg[bg>5]=5
    #img = 10*last_diff #(this_diff)#-last_diff) #
    img = 10*detect(this_diff,last_diff)
    shift = ensemblegetshift(this_noflash[0],last_noflash[0][0],searchbox=20,step=2,searchblocksize=50,ensemblesizesqrt=3)
    print(shift)
    #if dilate: noflash = getblockmaxedimage(noflash,blocksize,offset)
    res = alignandsubtract(last_diff,shift,this_diff,margin=10)
    #img = 10*res
    
    
    
    ##img[img>255]=255
    ##img+= 10*bg[10:-10,10:-10]
    
    #steps = 10
    #k = int(img.shape[0] / steps)
    #l = int(img.shape[1] / steps)
    #img = img[:k*steps,:l*steps].reshape(k,steps,l,steps).max(axis=(-1,-3))
    #plt.figure(figsize=[20,10])
    #plt.imshow(img,cmap='gray')
    #plt.clim([0,305])
    #plt.colorbar()
    #plt.hlines(truey/10-2,truex/10-12,truex/10-7,'w')
    #plt.hlines(truey/10-2,truex/10+3,truex/10+8,'w')
    #plt.vlines(truex/10-2,truey/10-12,truey/10-7,'w')
    #plt.vlines(truex/10-2,truey/10+3,truey/10+8,'w')
    
    
    #img = res.copy()
    img = this_flash[0]-this_noflash[0] #(this_noflash[0]+this_noflash[1])/2
    searchimg = res.copy()
    for i in range(10):
        y,x = unravel_index(searchimg.argmax(), searchimg.shape)
        
        alignmentcorrection = 20
        imgcorrection = 20
        #if (x<savesize) or (y<savesize) or (x>searchimg.shape[1]-savesize-1) or (y>searchimg.shape[0]-savesize-1): continue
        target = 1*(((y-truey+alignmentcorrection)**2 + (x-truex+alignmentcorrection)**2)<10**2)
        print(x,truex,y,truey)
        
        patch = img[y-savesize+imgcorrection:y+savesize+imgcorrection,x-savesize+imgcorrection:x+savesize+imgcorrection].astype(np.float32)
        patchn+=1
        searchimg[y-delsize:y+delsize,x-delsize:x+delsize]=0
        patches.append({'patch':patch,'target':target})
```
